We next check for the columns which contains `'T'` or `'M'`.

```python
missing = [column for column in weather.columns if ('M' in weather[column].astype(str).unique()) or ('-' in weather[column].astype(str).unique())]
weather=weather.replace('-', np.nan)
weather=weather.replace('M', np.nan)
```

We can immediately fill in the missing values of `tavg_y` by taking the mean of tmax and tmin.

```python
def fill_Tavg(x1,x2,x3):
    if pd.isnull(x3):
        x3=np.ceil((x1+x2)/2)
        return x3
    else:
        return x3
```

```python
weather['tavg_y']=weather.apply(lambda x: fill_Tavg(x['tmax_y'], x['tmin_y'], x['tavg_y']), axis=1)
```
